```python
import pandas as pd
from transformers import pipeline
from tqdm import tqdm
```

```python
all_emotion = pipeline("text-classification", model="MilaNLProc/xlm-emo-t", top_k=10)

all_emotion("Oggi sono proprio contento!")
```

```python
peak_emotion = pipeline("text-classification", model="MilaNLProc/xlm-emo-t", top_k=1)

peak_emotion("Oggi sono proprio contento!")[0][0]['label']
```

## 1. Italian Emotion detection

```python
url = "https://github.com/RacheleSprugnoli/Esercitazioni_SA/raw/master/dataset/MultiEmotions-It.tsv"
italian_df = pd.read_csv(url, sep='\t')
```

```python
# Initialize the pipeline
pipe = pipeline("text-classification", model="MilaNLProc/xlm-emo-t", top_k=1)

tqdm.pandas(desc="Emotion Inference")
italian_df['Inferenced_Emotion'] = italian_df['comment'].progress_apply(lambda x: pipe(x)[0][0]['label'])
```

```python
italian_df[['Inferenced_Emotion']].value_counts()
```

```python
file_name = 'italian_inferenced_emotion.csv'
italian_df.to_csv(file_name, index=False)

# Use Colab's files module to download
from google.colab import files
files.download(file_name)
```

## 2. German Emotion Detection

```python
url = "https://github.com/OweysMomenzada/Improving-Emotion-Detection-with-context-sensitive-classification-for-German-text-corpus/raw/main/Data%20collecting/fullset.csv"
german_df = pd.read_csv(url, sep=',')
german_df.drop(columns=["Unnamed: 0", "Text"], inplace=True)
```

```python
german_df.head()
```

```python
# Initialize the pipeline
pipe = pipeline("text-classification", model="MilaNLProc/xlm-emo-t", top_k=1)

tqdm.pandas(desc="Emotion Inference")
german_df['Inferenced_Emotion'] = german_df['text_de'].progress_apply(lambda x: pipe(x)[0][0]['label'])
```

```python
german_df[['Inferenced_Emotion']].value_counts()
```

```python
file_name = 'german_inferenced_emotion.csv'
german_df.to_csv(file_name, index=False)

# Use Colab's files module to download
from google.colab import files
files.download(file_name)
```

## 3. French Emotion detection

```python
url = "https://github.com/smnbrnrd/EmoDIFT/raw/main/emodift_manual_annot_2215.csv"
french_df = pd.read_csv(url, sep=',')
french_df.drop(columns=["num",], inplace=True)
# Mapping French column names to English equivalents
column_translation = {
    'Colere': 'Anger',
    'Degout': 'Disgust',
    'Peur': 'Fear',
    'Mefiance': 'Mistrust',
    'Tristesse': 'Sadness',
    'Surprise': 'Surprise',
    'Ironie': 'Irony',
    'Neutre': 'Neutral'
}

# Renaming the columns in the DataFrame
french_df.rename(columns=column_translation, inplace=True)

french_df['Emotion'] = french_df.iloc[:, 1:].idxmax(axis=1)
```

```python
french_df.head()
```

```python
# Initialize the pipeline
pipe = pipeline("text-classification", model="MilaNLProc/xlm-emo-t", top_k=1)

tqdm.pandas(desc="Emotion Inference")
french_df['Inferenced_Emotion'] = french_df['tweet'].progress_apply(lambda x: pipe(x)[0][0]['label'])
```

```python
french_df[['Inferenced_Emotion']].value_counts()
```

```python
file_name = 'french_inferenced_emotion.csv'
french_df.to_csv(file_name, index=False)

# Use Colab's files module to download
from google.colab import files
files.download(file_name)
```
